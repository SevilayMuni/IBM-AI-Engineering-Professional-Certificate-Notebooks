# Regression Model with KERAS for Concrete Strength

### Data Description:
Predictors: Cement, Blast Furnace Slag, Fly Ash, Water, Superplasticizer, Coarse Aggregate, Fine Aggregate

Target: Strength

```python
# Importing packages
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Loading data
data = pd.read_csv('concrete_data.csv')
data.head()
```

```python
data.shape
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# Basic stats for data
data.describe()
```

```python
# Separating target variable
X = data.drop('Strength', axis = 1).reset_index(drop = True)
y = data['Strength']
```

```python
# Checking dataframes
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.shape
```

## Part A:  *Baseline Model*

```python
column_n = X.shape[1]
column_n
```

```python
# Initializing the model
model_a = Sequential()

# Building the model
model_a.add(Dense(10, activation = 'relu', input_shape = (column_n,))) # 1 hidden layer with 10 nodes
model_a.add(Dense(1)) # outer layer
model_a.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Repeating training, predicting and calculating MSE btw prediction and y_test (true value) for 50 times

mse_list_a = []

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
    model_a.fit(X_train, y_train, epochs = 50, batch_size = 103, verbose = 1)
    preds_a = model_a.predict(X_test)
    mse_a = mean_squared_error(y_test, preds_a)
    mse_list_a.append(mse_a)
```

```python
# Checking calculations
mse_list_a
```

```python
print('For Part A:')
print('Mean MSE of 50 train/test rounds:', round(np.mean(mse_list_a), 3))
print('Standard Deviation (Std):', round(np.std(mse_list_a), 3))
```

## Part B:  *2nd Model with Normalized Predictor Variables*

```python
# Normalizing predictor variables
X = (X - X.mean()) / X.std()
X.head()
```

```python
# Initializing Part B model
model_b = Sequential()

# Building the model
model_b.add(Dense(10, activation = 'relu', input_shape = (column_n,))) # 1 hidden layer with 10 nodes
model_b.add(Dense(1)) # outer layer
model_b.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Repeating training, predicting and calculating MSE btw prediction and y_test (true value) for 50 times

mse_list_b = []

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
    model_a.fit(X_train, y_train, epochs = 50, batch_size = 103, verbose = 1)
    preds_b = model_a.predict(X_test)
    mse_b = mean_squared_error(y_test, preds_b)
    mse_list_b.append(mse_b)
```

```python
print('For Part B')
print('Mean MSE of 50 train/test rounds:', round(np.mean(mse_list_b), 3))
print('Standard Deviation (Std):', round(np.std(mse_list_b), 3))
```

## Part C: *3rd Model with Increased Number of Epochs*

```python
X.head()
```

```python
mse_list_c = []

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
    model_a.fit(X_train, y_train, epochs = 100, batch_size = 103, verbose = 1)
    preds_c = model_a.predict(X_test)
    mse_c = mean_squared_error(y_test, preds_c)
    mse_list_c.append(mse_c)
```

```python
print('For Part C')
print('Mean MSE of 50 train/test rounds:', round(np.mean(mse_list_c), 3))
print('Standard Deviation (Std):', round(np.std(mse_list_c), 3))
```

## Part D: *4th Model with More Hidden Layers* 

```python
# Initializing Part C model
model_d = Sequential()

# Building the model
model_d.add(Dense(10, activation = 'relu', input_shape = (column_n,))) # 1st hidden layer with 10 nodes
model_d.add(Dense(10, activation = 'relu')) # 2nd hidden layer with 10 nodes
model_d.add(Dense(10, activation = 'relu')) # 3rd hidden layer with 10 nodes
model_d.add(Dense(1)) # outer layer
model_d.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
mse_list_d = []

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
    model_d.fit(X_train, y_train, epochs = 50, batch_size = 103, verbose = 1)
    preds_d = model_d.predict(X_test)
    mse_d = mean_squared_error(y_test, preds_d)
    mse_list_d.append(mse_d)
```

```python
print('For Part D')
print('Mean MSE of 50 train/test rounds:', round(np.mean(mse_list_d), 3))
print('Standard Deviation (Std):', round(np.std(mse_list_d), 3))
```

## Result Table and Interpretation

### Result table derived from 4 different models:

```python
# Creating dataframe for the results of 4 models
# Part A model is called as 'Initial', Part B is refered w/ 'Normalized Predictors'
# Part C model -- '100 Epochs', Part D -- '3 Hidden Layers'

scores_list = [['Initial', 221.12, 402.182], ['Normalized Predictors', 87.489, 62.292], 
               ['100 Epochs', 38.664, 0.664], ['3 Hidden Layers', 37.852, 23.886]] 
scores_df = pd.DataFrame(scores_list, columns = ['Feature of Model','Mean of MSEs', 'Std of MSEs'])
scores_df
```

### Interpretation

1) Nomalization of predictive variables drastically lowered both mean and std compare to Part A aka 'Initial'.
   - Therefore, **normalization increased prediction accuracy** of the model and SMEs more tightly clustered around the mean.

2) Compare to Part B, training the model with more epochs further decreased mean of MSEs. It also diminished std to degree where there is almost no variance among MSEs.

3) Compare to Part B, model training with more hidden layers(3) substantially reduced mean of MSEs. It also lowered standard deviation. 
   - **Adding more hidden layers increased prediction accuracy** and SMEs more tightly clustered around the mean.
   
4) Part C and Part D models yielded almost same mean of MSEs; nevertheless, there is huge difference btw std values.
   - Very low std value (Part C) might be indicator of overfitting, since too many epochs cause the model memorizing training data instead of learning underlying patterns.
   

All in all, **variable normalization** and adding **more hidden layers** to model might be safe and sound strategy to **increase prediction accuracy** of regression model.
